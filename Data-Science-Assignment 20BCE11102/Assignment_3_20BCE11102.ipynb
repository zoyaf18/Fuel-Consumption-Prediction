```python
#Handling Outliers
```

```python
from scipy.stats import zscore
from scipy.stats import mstats

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
z_scores = zscore(df[numeric_columns])

# Set the threshold for outliers (e.g., 3 standard deviations)
threshold = 0.03

# Find outliers and replace them with the nearest value within the threshold
df[numeric_columns] = df[numeric_columns].apply(lambda x: mstats.winsorize(x, limits=(0, threshold)).data.flatten())

# Print the updated DataFrame with winsorized values
print(df.head())



# Box plot
plt.boxplot(df['price'])
plt.ylabel('Variable')
plt.title('Box plot of Variable')
plt.show()
```

```python
#Encoding

# Check for categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns)
```

```python
print(df_encoded.columns)
```

```python
# Splitting the data into independent and dependent variables

# Features or independent variables
X = df_encoded[['area', 'bedrooms', 'bathrooms', 'stories', 'parking',
       'mainroad_no', 'mainroad_yes', 'guestroom_no', 'guestroom_yes',
       'basement_no', 'basement_yes', 'hotwaterheating_no',
       'hotwaterheating_yes', 'airconditioning_no', 'airconditioning_yes',
       'furnishingstatus_furnished', 'furnishingstatus_semi-furnished',
       'furnishingstatus_unfurnished']]  

y = df_encoded['price']  # Target or dependent variable
```

```python
#Scale the independent variables

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Split the data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Model Building
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#train the model
lr.fit(X_train,y_train)
```

```python
#test the model 
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
print('R-squared(Accuracy):', r2)
```
